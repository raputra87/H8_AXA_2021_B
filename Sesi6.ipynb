## Bar Charts

```python
df_iceland = df_can.loc['Iceland', years]   #call out data iceland tiap years nya 
df_iceland.head() 
```

```python

df_iceland.plot(kind='bar',
               figsize=(10, 6))

plt.title('Icelandic immigrants to Canada from 80 to 13')
plt.ylabel('Years')
plt.xlabel('Number of Immigrants')

# Annotate arrow
plt.annotate('text',                      # s: str. will leave it blank for no text
             xy=(32, 70),             # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),         # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',         # will use the coordinate system of the object being annotated 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy=(28, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation=72.5,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',                      # want the text to be horizontally 'left' algned.
            )

plt.show()
```

## Pie Chart

```python
#dataframe berdasarkan continents 
df_continents = df_can.groupby('Continent', axis=0).sum()       #untuk axix 0 data nya di sum , jadinya pertahun di sum 
df_continents                                                   #untuk jadi df_continents 
```

```python
#membuat plot py untuk total imigran dari tiap data continents dari df_continents
df_continents['Total'].plot(kind='pie')
```

```python

df_continents['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%50f%%',      #add in percentages angel point
                            startangle = 45,        #startangle dari 90 derajat, jadinya bisa di lihat yg Africa
                            shadow=True
                           )
plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')       #set pied in chart lool like a circle 

plt.show()
```

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0.1, 0.1, 0.1]       #rasio for each continent

df_continents['Total'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%9.1f%%',
                            startangle = 90,
                            shadow=True,
                            colors=color_list,
                            explode=explode_list,
                            labels=None,
                            pctdistance=1.15
                           )
plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.15)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0.1, 0.1, 0.1]

df_continents['2013'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%',
                            startangle = 90,
                            shadow=True,
                            colors=color_list,
                            explode=explode_list,
                            labels=None,
                            pctdistance=1.15
                           )
plt.title('Immigration to Canada by Continent in 2013', y=1.15)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

## Subplots

```python
#subplots untuk membuat  2 plot dalam satu figure yang sama 
fig = plt.figure()  # create figure

ax0 = fig.add_subplot(1, 2, 1)      # create subplots(numerrow,number cols, plot_number ) ||brarti disini ada 1 row 2 col, plot ke 1
ax1 = fig.add_subplot(1, 2, 2)      # create subplots(numerrow,number cols, plot_number ) ||brarti disini ada 1 row 2 col, plot ke 2
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)


# First Plot
df_CI.plot(kind='box', color='salmon', vert=False, figsize=(20, 6), ax=ax0) #ada tambahan ax, ang untuk meneruskan ax0 karna berupa subplot
ax0.set_title('Box Plots of Immigrants from China and India')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Second Plot
df_CI.plot(kind='line', figsize=(20, 6), ax=ax1)
ax1.set_title('Line Plots of Immigrants from China and India')
ax1.set_xlabel('Number of Immigrants')
ax1.set_ylabel('Years')


# tird plot 
df_CI.plot(kind='line', figsize=(20, 6), ax=ax3)
ax3.set_title('empty chart')
ax3.set_xlabel('Number of Immigrants')
ax3.set_ylabel('Years')


# tird plot 
df_CI.plot(kind='line', figsize=(20, 6), ax=ax3)
ax4.set_title('empty chart')
ax4.set_xlabel('Number of Immigrants')
ax4.set_ylabel('Years')


plt.show()
```

## Scatter Plots

```python
#untuk 2 D membanding variable satu sama lain 


#SUM to get total pupulation per year
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

#change to integer yg ada di total 
df_tot.index = map(int, df_tot.index)

#reset index tup back in df_tot
df_tot.reset_index(inplace=True)

#rename jadi year sama total 
df_tot.columns = ['year', 'total']
```

```python
df_tot.head()
```

```python
# plotting as scatter kemudai kata kuncii untuk berapa di x dan y nya yg mana  
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='palevioletred')

plt.title('Total Immigration to Canada')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```
