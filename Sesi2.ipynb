Python break and continues statment

image.png

```python
# Break
n = 5

while n > 0:
    n -= 1        # n berkurang 1 
    if n == 2:
        break
    print(n)
print('Loop Ended')
```

```python
# Continue
n = 5

while n > 0:
    n -= 1
    if n == 2:
        continue
    print(n)
print('Loop Ended')
```

```python
# Else clause

n = 5

while n > 0:
    n -= 1
    print(n)
else:
    print('Loop Ended')
```

```python
n = 5

while n > 0:
    n -= 1
    if n == 2:
        break
    print(n)
else:
    print('Loop Ended')

# n> 0 menjadi false. Klausa else dieksekusi
```
